State/UT-wise Number of Indian Penal Code (IPC) Crimes from 2020 to 2022"

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("NCRB_Table_1A.1.csv")
```

```python
# this delete row where total of all colimns state,ut and all india 
df=df.drop(index=28)
df=df.drop(index=37)
df=df.drop(index=38)
```

```python
fig,axes=plt.subplots(1,3,figsize=(18,6))
df_sorted=df.sort_values(by="2022",ascending=False)
sns.barplot(data=df_sorted,x='State/UT',y='2022',ax=axes[0],palette=("red","blue"))
axes[0].set_title("Year 2022")
for bin in axes[0].containers:
    axes[0].bar_label(bin,rotation=90)
axes[0].set_xticklabels(axes[0].get_xticklabels(),rotation=90)

df_sorted=df.sort_values(by="2021",ascending=False)
sns.barplot(data=df_sorted,x='State/UT',y='2021',ax=axes[1],palette=("red","blue"))
axes[1].set_title("Year 2021")
for bin in axes[1].containers:
    axes[1].bar_label(bin,rotation=90)
axes[1].set_xticklabels(axes[1].get_xticklabels(),rotation=90)


df_sorted=df.sort_values(by="2020",ascending=False)
sns.barplot(data=df_sorted,x='State/UT',y='2020',ax=axes[2],palette=("red","blue"))
axes[2].set_title("Year 2020")
for bin in axes[2].containers:
    axes[2].bar_label(bin,rotation=90)
axes[2].set_xticklabels(axes[2].get_xticklabels(),rotation=90)
```

```python
df['change_2021_vs_2020']=df['2021']-df['2020']
df['change_2022_vs_2021']=df['2022']-df['2021']
```

```python
plt.figure(figsize=(18,15))
ax=sns.barplot(df,y='change_2021_vs_2020',x='State/UT')
plt.xticks(rotation=90,ha='right')
for bin in ax.containers:
    ax.bar_label(bin,rotation=90)

plt.figure(figsize=(18,10))
ax=sns.barplot(df,y='change_2022_vs_2021',x='State/UT')
plt.xticks(rotation=90,ha='right')
for bin in ax.containers:
    ax.bar_label(bin,rotation=90)
```

```python
def trend(row):
    if row['2020']>row['2021']>row['2022']:
        return 'Consistent decrease'
    elif row['2020']<row['2021']<row['2022']:
        return 'Consistent increase'
    else:
        return 'Fluctuating'
df['trend']=df.apply(trend,axis=1)

def tren(value):
    if value == 'Consistent decrease':
        return -1
    elif value == 'Consistent increase':
        return 1
    else:
        return 0

df['trend_to_num'] = df['trend'].apply(tren)

```

```python
sns.barplot(data=df,x='State/UT',y='trend_to_num')
plt.xticks(rotation=90)
plt.title('Consistent Increases-decreases or Fluctuating')    
```
